```
def del_file(file_name):
  if os.path.exists(file_name):
    os.remove(file_name)
  else:
    print("EBIR:: The file %s does not exist", file_name)
    
```

```
del_file("categories.json")
!scrapy crawl es_leroymerlin_category_map -o categories.json
```

```
with open('categories.json', 'r') as json_file:
  data = json.load(json_file)
   
categories = []
for each in data:
  if(each['category'][0]['name']=='Espejos de aumento para maquillaje'):
    categories.append(each)


with open('categories.json', 'w') as json_file:
  for each in categories:
    print(each)
    
    json_file.write("["+str(each).replace("'", '"') +"]")
```
